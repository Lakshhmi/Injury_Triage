```python
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_excel("Injury Triage 2022.xlsx")
df.head()
```

```python
cv=CountVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=cv.fit_transform(df['Description of Event'])
dtm
```

```python
#Import LDA from sklearn and define the number of clusters
from sklearn.decomposition import LatentDirichletAllocation
LDA=LatentDirichletAllocation(n_components=2,random_state=42)
LDA.fit(dtm)
```

```python
#Display few words and counts
import random
for i in range(10):
    random_word_id = random.randint(0,1345)
    print(cv.get_feature_names_out()[random_word_id])
```

```python
#Display the top 15 words in each 3 clusters we defined
for index,topic in enumerate(LDA.components_):
    print(f'The top 15 words for topic #{index}')
    print([cv.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results=LDA.transform(dtm)
df['Category'] = topic_results.argmax(axis=1)
df.head(10)
```

### Non-negative Matrix factorization(NMF)

```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#preprocessing using term frequency and inverser document frequency
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm = tfidf.fit_transform(df['Description of Event'])
```

```python
nmf_model = NMF(n_components=2,random_state=42)
nmf_model.fit(dtm)
print("record size :",len(tfidf.get_feature_names_out()))
```

```python
#Display few words and counts
import random
for i in range(10):
    random_word_id = random.randint(0,1345)
    print(tfidf.get_feature_names_out()[random_word_id])
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f'The top 15 words for topic #{index}')
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-25:]])
    print('\n')
```

```python
#Assign topics back to data frames
topic_results=nmf_model.transform(dtm)
df['Category'] = topic_results.argmax(axis=1)
df[['Description of Event','ACS Category','Category']]
```
